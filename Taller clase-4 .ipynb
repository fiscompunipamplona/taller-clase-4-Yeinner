```python
from numpy import*
x = linspace(0, 10, 1000)
y = sin(x)
plot(x,y)
show()
```

```python
from pylab import*
x = linspace(0, 10, 100)
y = sin(x)
plot(x,y,'k--')
y = cos(x)
plot(x,y,'g-')
ylim(-1.1, 1.1)
xlim(0., 10)
grid("on")
xlabel("Eje X")
ylabel("Eje Y")
show()
```
